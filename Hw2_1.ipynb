Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)" посилання

Необхідно виконати:

Вивести перші рядки таблиці за допомогою методу head

```python
import pandas as pd

url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"
data_born = pd.read_html(url, attrs={"class": "wikitable sortable"}, decimal=",", thousands=".")
print(data_born[0].head())
```

Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
df = data_born[0]
rows, columns = df.shape
print("Кількість рядків: ", rows)
print("Кількість cтовпців: ", columns)
```

Замініть у таблиці значення "—" на значення NaN

```python
from numpy import NaN
df.replace("—", NaN)
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
print(df.dtypes)
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—" - всі числові.

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
missing_val = df.isnull().sum()/ len(df)
print("Missing values in columns: ", missing_val)
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df = df.drop(df.index[-1])
print(df)
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df = df.fillna(df.mean(numeric_only=True))
print(df)
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні, остання колонка 2014, отже рахуємо для 2014

```python
avg_2014 = df['2014'].mean()
print(avg_2014)
regions_more_2014 = df[df['2014'] > avg_2014]['регіон'].tolist()
print(regions_more_2014)
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
region_max_2014 = df[df['2014']==df['2014'].max()]['регіон']
print(region_max_2014)
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році (2014 є тільки)

```python
%matplotlib inline
import matplotlib.pyplot as plt
data_2014 = df[['регіон', '2014']]
plt.bar(data_2014['регіон'], data_2014['2014'])
plt.xticks(rotation=90)
plt.xlabel('Регіон')
plt.ylabel('Народжуваність у 2014 році')
plt.title('Народжуваність за регіонами у 2014 році')
plt.show
```

Графіки народжуваності за областями України в 1950 та 2014 роках

```python
data_2014 = df[['регіон', '2014']]
data_1950 = df[['регіон', '1950']]
fig, axs = plt.subplots()
axs.plot(data_2014['регіон'], data_2014['2014'], label='народжуваність у 2014 році')
axs.plot(data_1950['регіон'], data_1950['1950'], label='народжуваність у 1950 році')
plt.xticks(rotation=90)
plt.xlabel('Регіон')
plt.ylabel('Народжуваність у граничних роках')
plt.title('Народжуваність за регіонами')
plt.legend()
```

Оцінимо як змінювалася народжуваність у м. Київ

```python
kyiv_data = df[df['регіон'] == 'Київ']
data_without_region = kyiv_data.iloc[:, 1:]
years = data_without_region.columns
kyiv_birth_rate = data_without_region.loc[data_without_region.index[-1], :]

plt.plot(years, kyiv_birth_rate, color='green')
plt.xlabel('Рік')
plt.ylabel('Народжуваність')
plt.title('Зміна народжуваності в місті Київ')
plt.xticks(rotation=45)
plt.show()
```
