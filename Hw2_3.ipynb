```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
books['price'].plot(kind='hist')
plt.xlabel('Ціна')
plt.ylabel('Кількість')
plt.title('Розподіл цін')
plt.show()
print("Max price", books['price'].max())
print("Min price", books['price'].min())
print("Mean price", books['price'].mean())
print("Median price", books['price'].median())
```

Відповідь: Максимальна ціна? **105**
Відповідь: Мінімальна ціна? **0**
Відповідь: Середня ціна? **13.1**
Відповідь: Медіанна ціна? **11.0**

```python
rating_max = (books['user_rating'].max())
print("Max rating: ", rating_max)
count_rating_max = books['user_rating'].value_counts()[rating_max]
print("Count of books with max rating: ", count_rating_max)
review_max_book = books[books['reviews'] == books['reviews'].max()]['name']
print("Max reviewed book: ", review_max_book)
books_2015 = books[books['year'] == 2015]
expensive_2015_book = books_2015[books_2015['price'] == books_2015['price'].max()]['name']
print("Most expensive book in 2015: ", expensive_2015_book)
fiction_book_2010 = books[(books['year'] == 2010) & (books['genre'] == 'Fiction')]
count_fiction_book_2010 = len(fiction_book_2010)
print("Count of fiction books in 2010: ", count_fiction_book_2010)
rating_49_2010_2011 = books[(books['user_rating']==4.9)&((books['year']==2011)|(books['year']==2011))]
count_rating_49 = len(rating_49_2010_2011)
print("Count of books with rating 4.9 in 2010, 2011: ", count_rating_49)
cheap_books_2015 = books[(books['year'] == 2015) & (books['price'] < 8)]
sorted_books_2015 = cheap_books_2015.sort_values('price')
print("Sorted books cheaper then 8 in 2015: ", sorted_books_2015)
```

І насамкінець, давайте відсортуємо за зростанням ціни всі книги, які потрапили до рейтингу в 2015 році і коштують дешевше за 8 доларів (використовуйте функцію sort_values).
 Яка книга остання у відсортованому списку? Відповідь: **Old School (Diary of a Wimpy Kid #10)**  

```python
genre_prices = books.groupby('genre')['price'].agg(['min', 'max'])
print(genre_prices)
```

Відповідь: Максимальна ціна для жанру Fiction: **82**

Відповідь: Мінімальна ціна для жанру Fiction: **0**

Відповідь: Максимальна ціна для жанру Non Fiction: **105**

Відповідь: Мінімальна ціна для жанру Non Fiction: **0**



```python
author_count = books.groupby('author').agg(books_count=('name', 'count')).sort_values(by='books_count', ascending=False)
print(author_count)
```

Відповідь: Якої розмірності вийшла таблиця? **248x1**

Відповідь: Який автор має найбільше книг? **Jeff Kinney**

Відповідь: Скільки книг цього автора? **12**

```python
author_rating = books.groupby('author').agg(author_rating=('user_rating','mean')).sort_values(by='author_rating', ascending=False)
print(author_rating)
```

Відповідь: У якого автора середній рейтинг мінімальний?  **Donna Tartt**

Відповідь: Який у цього автора середній рейтинг? **3.9**

```python
from pandas.core.internals import concat
authors_info = pd.concat([author_rating, author_count], axis=1).sort_values(by=['books_count', 'author_rating'])
print(authors_info)
```

Відповідь: Який автор перший у списку?  **Muriel Barbery**

```python
sns.set_style('whitegrid')
sns.lineplot(x='year', y='price', hue='genre', data=books)
plt.title('Середня ціна книг за роками')
plt.xlabel('Рік')
plt.ylabel('Ціна, $')
plt.legend(title='Жанр')
```
